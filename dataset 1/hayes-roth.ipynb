```python
# import libraries
import pandas as pd
import numpy as np
import sklearn as sns
```

```python
# load data

"""
Attribute Information:
      -- 1. name: distinct for each instance and represented numerically
      -- 2. hobby: nominal values ranging between 1 and 3
      -- 3. age: nominal values ranging between 1 and 4
      -- 4. educational level: nominal values ranging between 1 and 4
      -- 5. marital status: nominal values ranging between 1 and 4
      -- 6. class: nominal value between 1 and 3
"""

df=pd.read_csv("hayes-roth.csv",names=['name','hobby','age','educational_level','martial_status','class'])
```

```python
# show top 5 rows of the dataset
df.head()
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
# shape of all features
x.shape
```

```python
# show features
x.head()
```

```python
# shape of all target label
y.shape
```

```python
# show target
y.head()
```

# preprocessing 

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x=scale.fit_transform(x)
```

```python
x
```

# K Fold Validation

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=9)
for train_index, test_index in kf.split(x):
    X_train, X_test = x[train_index], x[test_index]
    y_train, y_test = y[train_index], y[test_index]
```

```python
X_train.shape
```

```python
X_test.shape
```

# cross validation

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
score=cross_val_score(model,X_train,y_train,scoring="neg_mean_absolute_error")
```

# Mean Absolute Error

```python
print("Mean absolute Error {}".format(np.mean(abs(score))))
```

# Root Mean squre Error

```python
print("Root Mean Squre Error {}".format(np.sqrt(np.mean(abs(score)))))
```
